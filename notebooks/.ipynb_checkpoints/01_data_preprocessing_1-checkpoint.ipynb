# Data Preprocessing 

```python
import pandas as pd
import matplotlib.pyplot as plt
import mplfinance as mpf

import sys
sys.path.append('../src/')

from preprocessing import *
from utils import *
```

```python
#Load data from mt5

#EURUSD currency pair from 4.1.21 00:00 - 29.12.23 11:45
eur_usd_data = load_mt5_data('../data/raw/EURUSD_M15_202101040000_202312292345.csv')
#USDJPY currency pair from 4.1.21 00:00 - 29.12.23 11:45
usd_jpy_data = load_mt5_data('../data/raw/USDJPY_M15_202101040000_202312292345.csv')
#AUDUSD currency pair from 4.1.21 00:00 - 29.12.23 11:45
aud_usd_data = load_mt5_data('../data/raw/AUDUSD_M15_202101040000_202312292345.csv')
#GBPUSD currency pair from 4.1.21 00:00 - 29.12.23 11:45
gbp_usd_data = load_mt5_data('../data/raw/GBPUSD_M15_202101040000_202312292345.csv')
#NZDUSD currency pair from 4.1.21 00:00 - 29.12.23 11:45
nzd_usd_data = load_mt5_data('../data/raw/NZDUSD_M15_202101040000_202312292345.csv')
#USDCAD currency pair from 4.1.21 00:00 - 29.12.23 11:45
usd_cad_data = load_mt5_data('../data/raw/USDCAD_M15_202101040000_202312292345.csv')
#USDCHF currency pair from 4.1.21 00:00 - 29.12.23 11:45
usd_chf_data = load_mt5_data('../data/raw/USDCHF_M15_202101040000_202312292345.csv')
```

```python
#set index and drop unused columns
currency_pairs = [eur_usd_data, usd_jpy_data, aud_usd_data, gbp_usd_data, nzd_usd_data, usd_cad_data, usd_chf_data]
col = ['Date', 'Time', 'TickVol', 'Vol', 'Spread']

for pair in currency_pairs:
    set_datetime_index(pair)
    drop_col(pair, col)
```

```python
#data snippet
for data in currency_pairs:
    print(data.head())
```

```python
#Visualize Closing Price Data over time period

#date against closing price
titles = ['EUR_USD','USD_JPY','AUD_USD','GBP_USD','NZD_USD','USD_CAD','USD_CHF']
plot_close_price(currency_pairs, titles, nrows=len(titles), ncols=1, figsize=(15,20) )
```

```python
#Visualize candlestick chart of OHLC data in a day

# Plotting ohlc candlestick data
plot_ohlc_candle(currency_pairs, titles, nrows=len(titles), figsize=(20,40))
```

```python
#Clean and add features, set target val
for pair in currency_pairs:
    clean_data(pair)
    feature_engineering(pair)
    set_target(pair)
```

```python
#convert data to stationary (proven in EDA)
non_stationary_features = ['Open','High','Low','Close','EMA_5','EMA_10','EMA_20']
for pair in currency_pairs:
    convert_stationary_features(non_stationary_features, pair)
```
