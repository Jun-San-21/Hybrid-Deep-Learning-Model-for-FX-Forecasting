### Augmented Dickey-Fuller Test (ADF)

T = ∞: −3.43,
Significane level = 0.05

- null hypothesis : : non-stationary, where p-value > SL, ADF stat > T
else,

- reject null : : stationary


```python
from statsmodels.tsa.stattools import adfuller
pair = eur_usd_data
features = ['Open','High','Low','Close','EMA_5','EMA_10','EMA_20','RSI','Open_s','High_s','Low_s','Close_s','EMA_5_s','EMA_10_s','EMA_20_s']
adf_stats = []
p_values = []

for feature in features:
    result = adfuller(pair[feature])
    adf_stats.append(round(result[0], 2))
    p_values.append(round(result[1], 2))
print("            Open  High  Low  Close  EMA_5  EMA_10  EMA_20  RSI")
print(f'ADF stat: {adf_stats[:8]}')
print(f'p-value:  {p_values[:8]}')
print("\n")
print("            Open_s  High_s  Low_s  Close_s  EMA_5_s  EMA_10_s  EMA_20_s")
print(f'ADF stat: {adf_stats[8:]}')
print(f'p-value:  {p_values[8:]}')
```

##### result
null hypothesis: rejected for RSI, Open_s, High_s, Low_s, Close_s, EMA_s for 5, 10 and 20

## RSI to Close Price

```python
bins = [0, 30, 70, 100]
labels = ['Below 30', '30-70', 'Above 70']
eur_usd_data['RSI_Category'] = pd.cut(eur_usd_data['RSI'], bins=bins, labels=labels, include_lowest=True)

rsi_price_up_correlation = eur_usd_data.groupby('RSI_Category')['Target'].mean()
print(rsi_price_up_correlation)

plt.figure(figsize=(8, 5))
sb.barplot(x=rsi_price_up_correlation.index, y=rsi_price_up_correlation.values)
plt.title('Likelihood of Close Price Going Up by RSI Category')
plt.xlabel('RSI Category')
plt.ylabel('Proportion of Price Up (1)')
plt.show()

from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(eur_usd_data['RSI_Category'], eur_usd_data['Target'])

#chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")
```
